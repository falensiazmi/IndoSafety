```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
models = [
    'Sailor2-8B-Chat',
    'Sailor2-8B-Chat-FT',
    'SeaLLMs-v3-7B-Chat',
    'SeaLLMs-v3-7B-Chat-FT',
    'Llama-SEA-LION-v3-8B-IT',
    'Llama-SEA-LION-v3-8B-IT-FT',
    # 'Qwen2.5-0.5B-Instruct',
    # 'Qwen2.5-0.5B-Instruct-FT'
]
```

```python
DATANAME = 'indocloze'
print('Model, Average')
for model in models:
    try:
        df = pd.read_csv(f'results_downstream/output_{DATANAME}/result_prompt_{model}.csv')
    except:
        continue
    preds = df['preds'].tolist()
    preds = [{'A':0, 'B':1}[x] for x in preds]
    golds = df['answers'].tolist()
    print(model, end=', ')
    print(f"{accuracy_score(preds, golds)*100:.1f}")
```
